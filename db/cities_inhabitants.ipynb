```python
import pandas as pd

input_file = "cities_inhabitants.csv"

cities_df = pd.read_csv(input_file)
cities_df
```

```python
import requests

for ind, row in cities_df.iterrows():
    city = row["Name"]
    country = row["Member state"]
    
    api_url = "https://api.api-ninjas.com/v1/geocoding?city={city}&country={country}".format(city=city, country=country)

    # Read API key from local file
    with open("API_key.txt", "r") as file:
        api_key = file.read().strip()
    
    response = requests.get(api_url, headers={"X-Api-Key": api_key})

    if response.status_code == requests.codes.ok:
        data = response.json()
        if data[0]["name"] != city: print(data[0]["name"] + " != " + city) 

        cities_df.at[ind, "latitude"] = data[0]["latitude"]
        cities_df.at[ind, "longitude"] = data[0]["longitude"]
    else:
        print("Error:", response.status_code, response.text)

    print(cities_df.head())
    print()
```
